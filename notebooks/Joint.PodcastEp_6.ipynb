```python
import pandas as pd
import string

# Load the CSV file
df = pd.read_csv("Joint.PodcastEp_6.csv")

# Define a function to process the text: lowercase and remove punctuation
def preprocess_text(text):
    if isinstance(text, str):  # Ensure it's a string
        text = text.lower()  # Convert to lowercase
        text = text.translate(str.maketrans("", "", string.punctuation))  # Remove punctuation
    return text

# Apply the function to all string columns in the DataFrame
for column in df.select_dtypes(include=["object"]).columns:
    df[column] = df[column].apply(preprocess_text)

# Display the processed DataFrame
print(df.head())

# Save the cleaned data to a new file if needed
df.to_csv("Processed_PodcastEp_6.csv", index=False)
```

```python
from jiwer import wer

# Create a new column for WER
df["WER"] = df.apply(
    lambda row: wer(row["Corrected Transcription"], row["ASR Transcription"]), axis=1
)

# Display the updated DataFrame
print(df.head())
```

```python
average_wer = df["WER"].mean()
print(f"Average WER: {average_wer:.2f}")
```

```python
# Access the specific row containing all sentences
row_index = 0  # Adjust to the row index you want to analyze
specific_row = df.iloc[row_index]

# Split transcriptions into individual sentences
asr_sentences = specific_row["ASR Transcription"].split(". ")
corrected_sentences = specific_row["Corrected Transcription"].split(". ")

# Ensure both lists have the same length (handle cases with missing trailing periods)
min_length = min(len(asr_sentences), len(corrected_sentences))
asr_sentences = asr_sentences[:min_length]
corrected_sentences = corrected_sentences[:min_length]

# Filter and display non-matching sentences with their indices
non_matching_sentences = []
for i, (asr, corrected) in enumerate(zip(asr_sentences, corrected_sentences)):
    if asr != corrected:
        non_matching_sentences.append((i, asr, corrected))

# Display results
print("Non-matching sentences:")
for index, asr, corrected in non_matching_sentences:
    print(f"Sentence {index + 1}:")
    print(f"  ASR: {asr}")
    print(f"  Corrected: {corrected}")

```

```python
import pandas as pd

# Load your CSV file
df = pd.read_csv("Processed_PodcastEp_6.csv")

def compare_transcriptions(row):
    """
    Compares two transcriptions word by word to find added and removed words.

    Args:
        row: A row from the DataFrame containing 'ASR Transcription' and 'Corrected Transcription'.

    Returns:
        A dictionary with added and removed words.
    """
    asr_words = set(row["ASR Transcription"].split())
    corrected_words = set(row["Corrected Transcription"].split())

    added_words = corrected_words - asr_words
    removed_words = asr_words - corrected_words

    return {"added": added_words, "removed": removed_words}

# Get the first row
first_row = df.iloc[0]

# Compare the transcriptions
comparison_result = compare_transcriptions(first_row)

# Display the results
print("Words added in the corrected transcription:")
print(comparison_result["added"])
print("\nWords removed from the ASR transcription:")
print(comparison_result["removed"])
```
